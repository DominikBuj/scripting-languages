## Zadanie 0 (0 pkt)

Wczytaj do `DataFrame` plik CSV z narodzinami dzieci w Polsce:
`Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv`.

````python
import pandas as pd
data_frame = pd.read_csv('Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv', encoding='ISO-8859-1', sep=';')
display(data_frame)
````

## Zadanie 1
WyÅ›wietl najpopularniejsze imiÄ™ (i liczbÄ™) dla kaÅ¼dego rocznika.

````python
data_frame.loc[data_frame.groupby('Rok')['Liczba'].idxmax()][['Rok', 'Imie', 'Liczba']]
````

## Zadanie 2
WyÅ›wietl, ktÃ³re imiÄ™ dziewczynki i ktÃ³re chÅ‚opca jest najbardziej popularne Å‚Ä…cznie (kierujemy siÄ™
sumÄ… wystÄ…pieÅ„),

````python
girls = data_frame[data_frame["Plec"] == "K"]
boys = data_frame[data_frame["Plec"] == "M"]
print(f'najpopularniejsze imiÄ™ dziewczynki: {girls.groupby("Imie")["Liczba"].sum().idxmax()}')
print(f'najpopularniejsze imiÄ™ chÅ‚opca: {boys.groupby("Imie")["Liczba"].sum().idxmax()}')
````

## Zadanie 3
WyÅ›wietl, ktÃ³re imiona (pÅ‚eÄ‡ dowolna) wystÄ…piÅ‚y co najmniej 1000 razy rocznie w co najmniej 10 latach.

````python
(data_frame[data_frame['Liczba'] >= 1000].groupby('Imie').count() >= 10).index.to_list()
````

## Zadanie 4
WyÅ›wietl:
* w ktÃ³rym roku chÅ‚opcy mieli najwyÅ¼szy odsetek urodzin (np. 51.7%) 
* i w ktÃ³rym roku dziewczynki. 

Podaj te odsetki.

````python
name_count_by_year = data_frame.groupby('Rok')['Liczba'].sum()
print('name count by year')
display(name_count_by_year)

name_ratio_by_year = pd.DataFrame({
    'M': data_frame[data_frame['Plec'] == 'M'].groupby('Rok')['Liczba'].sum() / name_count_by_year,
    'K': data_frame[data_frame['Plec'] == 'K'].groupby('Rok')['Liczba'].sum() / name_count_by_year
})
print('name ratio by year\n')
print(name_ratio_by_year)

print(f'\nnajwyÅ¼szy odsetek chÅ‚opcÃ³w: {round(100 * name_ratio_by_year["M"].max(), 2)}%, w roku: {name_ratio_by_year["M"].idxmax()}')
print(f'najwyÅ¼szy odsetek dziewczynek: {round(100 * name_ratio_by_year["K"].max(), 2)}%, w roku: {name_ratio_by_year["K"].idxmax()}')
````

## Zadanie 5

WyÅ›wietl ile imion (pÅ‚eÄ‡ dowolna) z roku 2017 wystÄ…piÅ‚o po raz pierwszy.

````python
unique_names_not_year_2017 = data_frame[data_frame['Rok'] != 2017]['Imie'].unique()
year_2017 = data_frame[data_frame['Rok'] == 2017]
unique_names_year_2017 = year_2017[~year_2017['Imie'].isin(unique_names_not_year_2017)]['Imie']
print(f'iloÅ›Ä‡ imion, ktÃ³re wystÄ…piÅ‚y po raz pierwszy w roku 2017: {unique_names_year_2017.count()}')
````
