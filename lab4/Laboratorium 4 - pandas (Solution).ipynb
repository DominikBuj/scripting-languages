## Zadanie 0 (0 pkt)

Wczytaj do `DataFrame` plik CSV z narodzinami dzieci w Polsce:
`Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv`.

````python
import pandas as pd
df = pd.read_csv('Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv', encoding='ISO-8859-2', sep=';')
df = df.sort_values('Rok')
````

## Zadanie 1
Wy≈õwietl najpopularniejsze imiƒô (i liczbƒô) dla ka≈ºdego rocznika.

````python
df.loc[df.groupby('Rok')['Liczba'].idxmax()][['Rok', 'Imie', 'Liczba']]
````

## Zadanie 2
Wy≈õwietl, kt√≥re imiƒô dziewczynki i kt√≥re ch≈Çopca jest najbardziej popularne ≈ÇƒÖcznie (kierujemy siƒô
sumƒÖ wystƒÖpie≈Ñ),

````python
print(f"Najbardziej popularne imiƒô dziewczynki: {df[df['Plec'] == 'K'].groupby('Imie')['Liczba'].sum().idxmax()}")
print(f"Najbardziej popularne imiƒô ch≈Çopca    : {df[df['Plec'] == 'M'].groupby('Imie')['Liczba'].sum().idxmax()}")
````

## Zadanie 3
Wy≈õwietl, kt√≥re imiona (p≈Çeƒá dowolna) wystƒÖpi≈Çy co najmniej 1000 razy rocznie w co najmniej 10 latach.

````python
(df[df['Liczba'] > 1000].groupby('Imie').count() >= 10).index.to_list()
````

## Zadanie 4
Wy≈õwietl:
* w kt√≥rym roku ch≈Çopcy mieli najwy≈ºszy odsetek urodzin (np. 51.7%) 
* i w kt√≥rym roku dziewczynki. 

Podaj te odsetki.

````python
df3_1 = df.groupby('Rok')['Liczba'].sum()
df3_2 = pd.DataFrame({
    'M': df[df['Plec'] == 'M'].groupby('Rok')['Liczba'].sum() / df3_1,
    'K': df[df['Plec'] == 'K'].groupby('Rok')['Liczba'].sum() / df3_1
})
print(df3_2)
print("\nOdpowiedzi:")
print(f"Ch≈Çopcy mieli najwy≈ºszy odsetek ({round(100*df3_2['M'].max(), 2)}%) urodzin w roku {df3_2['M'].idxmax()}.")
print(f"Dziewczynki mia≈Çy najwy≈ºszy odsetek ({round(100*df3_2['K'].max(), 2)}%) urodzin w roku {df3_2['K'].idxmax()}.")
````

## Zadanie 5

Wy≈õwietl ile imion (p≈Çeƒá dowolna) z roku 2017 wystƒÖpi≈Ço po raz pierwszy.

````python
nie_2017 = df[df['Rok'] != 2017]['Imie'].unique()
df_2017 = df[df['Rok'] == 2017]
ile = df_2017[~df_2017['Imie'].isin(nie_2017)]['Imie'].count()
print(f'W 2017 roku pojawi≈Ço siƒô {ile} nowych imion.')
````
