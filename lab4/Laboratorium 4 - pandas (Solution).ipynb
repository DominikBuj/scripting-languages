## Zadanie 0 (0 pkt)

Wczytaj do `DataFrame` plik CSV z narodzinami dzieci w Polsce:
`Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv`.

````python
import pandas as pd
df = pd.read_csv('Najpopularniejsze-imiona-w-Polsce-w-latach-2000-2017.csv', encoding='ISO-8859-2', sep=';')
df = df.sort_values('Rok')
````

## Zadanie 1
Wywietl najpopularniejsze imi (i liczb) dla ka偶dego rocznika.

````python
df.loc[df.groupby('Rok')['Liczba'].idxmax()][['Rok', 'Imie', 'Liczba']]
````

## Zadanie 2
Wywietl, kt贸re imi dziewczynki i kt贸re chopca jest najbardziej popularne cznie (kierujemy si
sum wystpie),

````python
print(f"Najbardziej popularne imi dziewczynki: {df[df['Plec'] == 'K'].groupby('Imie')['Liczba'].sum().idxmax()}")
print(f"Najbardziej popularne imi chopca    : {df[df['Plec'] == 'M'].groupby('Imie')['Liczba'].sum().idxmax()}")
````

## Zadanie 3
Wywietl, kt贸re imiona (pe dowolna) wystpiy co najmniej 1000 razy rocznie w co najmniej 10 latach.

````python
(df[df['Liczba'] > 1000].groupby('Imie').count() >= 10).index.to_list()
````

## Zadanie 4
Wywietl:
* w kt贸rym roku chopcy mieli najwy偶szy odsetek urodzin (np. 51.7%) 
* i w kt贸rym roku dziewczynki. 

Podaj te odsetki.

````python
df3_1 = df.groupby('Rok')['Liczba'].sum()
df3_2 = pd.DataFrame({
    'M': df[df['Plec'] == 'M'].groupby('Rok')['Liczba'].sum() / df3_1,
    'K': df[df['Plec'] == 'K'].groupby('Rok')['Liczba'].sum() / df3_1
})
print(df3_2)
print("\nOdpowiedzi:")
print(f"Chopcy mieli najwy偶szy odsetek ({round(100*df3_2['M'].max(), 2)}%) urodzin w roku {df3_2['M'].idxmax()}.")
print(f"Dziewczynki miay najwy偶szy odsetek ({round(100*df3_2['K'].max(), 2)}%) urodzin w roku {df3_2['K'].idxmax()}.")
````

## Zadanie 5

Wywietl ile imion (pe dowolna) z roku 2017 wystpio po raz pierwszy.

````python
nie_2017 = df[df['Rok'] != 2017]['Imie'].unique()
df_2017 = df[df['Rok'] == 2017]
ile = df_2017[~df_2017['Imie'].isin(nie_2017)]['Imie'].count()
print(f'W 2017 roku pojawio si {ile} nowych imion.')
````
