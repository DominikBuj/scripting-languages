### Zadanie 1

Pobierz plik `english.50MB` z adresu http://pizzachili.dcc.uchile.cl/texts/nlang/ a następnie rozpakuj.

Skonwertuj litery na małe, usuń znaki interpunkcyjne i białe znaki (tabulatory, znaki nowej linii, wiele spacji zamień na pojedynczą). Wynik zapisz w pliku `english.txt`.

**Wskazówka**: Aby otworzyć plik możesz użyć poniższego fragmentu kodu:

````python
import re
with open("english.50MB", "r", encoding="latin-1") as input_file:
    text = input_file.read().lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    with open("english.txt", "w", encoding="latin-1") as output_file:
        output_file.write(text)
        print(f"word count: {len(text.split())}")
````

### Zadanie 2

Znajdź wszystkie słowa w pliku `english.txt`, które są palindromami.

#### Co to jest palindrom?
Palindrom to ciąg znaków $s$, który czytany od końca jest tym samym ciągiem znaków $s_0, s_1, \cdots, s_{n-1} = s_{n-1}, \cdots, s_1, s_0$, np. `aaa`, `abba`, `abcdedcba`, `aba`, `xyzyx`.

#### Jak wykonać zadanie?
Kroki:
* Krok 1: Odczytaj wszystkie unnikalne słowa z pliku (możesz umieścić je w zbiorze).
* Krok 2: Jeżeli `s == s[::-1]`, to znaczy, że słowo `s` jest palindromem.
* Krok 3: Wszystkie palindromy umieść w zbiorze (`set`).
* Krok 4: Wyświetl wszystkie palindromy.

````python
with open("english.txt", "r", encoding="latin-1") as input_file:
    unique_palindromes = set()
    for word in set(input_file.read().split()):
        if word == word[::-1]:
            unique_palindromes.add(word)
    print(unique_palindromes)
    print(f"unique palindromes count: {len(unique_palindromes)}")
````

### Zadanie 3

Znajdź anagramy w pliku `english.txt`.

#### Co to jest anagram?
Anagramem słowa $s$ jest inne słowo utworzone poprzez przestawienie jego liter. Np. `pakt` i `ptak`.

#### Jak wykonać zadanie?
* Krok 1: Umieść wszystkie unikalne słowa składające się z więcej niż jednego znaku w zbiorze.
* Krok 2: Utwórz słownik, gdzie kluczem są posortowane litery słowa, a wartością jest lista wszystkich słów (z pliku `english.txt`), których posortowane litery odpowiadają kluczowi (są takim samym ciągiem znaków).
* Krok 3: Wyświetl wszystkie elementy słownika, które posiadają więcej niż jedno słowo na liście.

Na przykład dla słów: `alga`, `pakt`, `gala`, `ptak`, `lata`, wynikowy słownik powinien wyglądać następująco:
```python
{
    'aagl': ['alga', 'gala'],
    'akpt': ['pakt', 'ptak'],
    # 'aalt': ['lata'] tego NIE wyświetlamy
}
```

````python
with open("english.txt", "r", encoding="latin-1") as input_file:
    unique_words = set([word for word in input_file.read().split() if len(word) > 1])
    anagrams = {}
    for word in unique_words:
        sorted_word = ''.join(sorted(word))
        anagrams[sorted_word] = anagrams[sorted_word] + [word] if sorted_word in anagrams else [word]
    anagrams = {key: value for key, value in anagrams.items() if len(anagrams[key]) > 1}
    for anagram in anagrams:
        print(anagram, anagrams[anagram])
    print(f"anagrams count: {len(anagrams)}")
````

3. Z poziomu skryptu: wczytaj wszystkie akapity z pliku, każdy do osobnego stringu (np. na liście).
Zbuduj 100 „losowych” funkcji haszujących. 
> Sugestia: funkcją „bazową” jest po prostu `hash(...)`. Zakładamy 64-bitową wersję Pythona 3.x, wtedy `hash(...)` jest 64-bitowy.
Do listy `seeds` umieszczamy 100 losowych liczb 64-bitowych.
`i`-ty hash dla ciągu `s` to będzie `hash(s) ^ seeds[i]` (użycie operatora XOR).

4. Przyjmij niewielką wartość `Q` (np. 15).
Dla każdego ciągu (tj. rozważanego akapitu):
* oznacz jego długość przez `n`,
* dla każdej ze 100 funkcji haszujących policz hasza w przesuwnym oknie tekstu o długości `Q` znaków (czyli łącznie mamy `n - Q + 1` wartości hasza); zapamiętaj MINIMUM z tych `n - Q + 1` wartości.

> Na wyjściu mamy zatem (dla 11 akapitów) 11 * 100 wartości haszy.

5. Rozważ pary akapitów „każdy z każdym”. Jeśli dla danej pary co najmniej (np.) 30 haszy jest wspólnych, to uważamy akapity za podobne (być może plagiat) i wyświetlamy na ekranie.

6. Wyświetl czas obliczeń (powinien być <0.5s).

7. Poeksperymentuj z liczbą użytych funkcji haszujących, wartością `Q`, stopniem modyfikacji oryginalnych akapitów tekstu, progiem detekcji akapitów podobnych.

````python
import random
import numpy as np
import itertools
import time

def get_file_lines(file):
    with open(file, "r", encoding="utf-8") as input_file:
        lines = [line.rstrip() for line in input_file.readlines()]
        return [line for line in lines if line]

def get_min_hash(line, seed, q):
    hashes = []
    for end in range(q - 1, len(line)):
        start = end - q + 1
        hashes.append(hash(line[start:end + 1]) ^ seed)
    return min(hashes)

def print_similar_lines(file, seeds_count, q, limit):
    lines = get_file_lines(file)
    seeds = [random.getrandbits(64) for i in range(seeds_count)]
    lines_hashes = {line:np.array([get_min_hash(line, seed, q) for seed in seeds]) for line in lines}
    similar_lines = {}
    for line_hashes, compared_line_hashes in itertools.permutations(lines_hashes, 2):
        if len(np.intersect1d(lines_hashes[line_hashes], lines_hashes[compared_line_hashes])) >= limit:
            similar_lines[line_hashes] = similar_lines[line_hashes] + [compared_line_hashes] if line_hashes in similar_lines else [compared_line_hashes]
    for line in similar_lines:
        print(line)
        for similar_line in similar_lines[line]:
            print(similar_line)
        print()

file = "wikipedia.txt"
seeds_count = 100
q = 10
limit = 15

start = time.time()
print_similar_lines(file, seeds_count, q, limit)
end = time.time()
print(f"execution time: {(end - start):.2f} seconds")
````
