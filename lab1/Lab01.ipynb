### Zadanie 1

Pobierz plik `english.50MB` z adresu http://pizzachili.dcc.uchile.cl/texts/nlang/ a nastpnie rozpakuj.

Skonwertuj litery na mae, usu znaki interpunkcyjne i biae znaki (tabulatory, znaki nowej linii, wiele spacji zamie na pojedyncz). Wynik zapisz w pliku `english.txt`.

### Zadanie 2

Znajd藕 wszystkie sowa w pliku `english.txt`, kt贸re s palindromami.

#### Co to jest palindrom?
Palindrom to cig znak贸w $s$, kt贸ry czytany od koca jest tym samym cigiem znak贸w $s_0, s_1, \cdots, s_{n-1} = s_{n-1}, \cdots, s_1, s_0$, np. `aaa`, `abba`, `abcdedcba`, `aba`, `xyzyx`.

#### Jak wykona zadanie?
Kroki:
* Krok 1: Odczytaj wszystkie unnikalne sowa z pliku (mo偶esz umieci je w zbiorze).
* Krok 2: Je偶eli `s == s[::-1]`, to znaczy, 偶e sowo `s` jest palindromem.
* Krok 3: Wszystkie palindromy umie w zbiorze (`set`).
* Krok 4: Wywietl wszystkie palindromy.

````python
with open("english.txt", "r", encoding="latin-1") as input_file:
    unique_palindromes = set()
    for word in set(input_file.read().split()):
        if word == word[::-1]:
            unique_palindromes.add(word)
    print(unique_palindromes)
    print(f"unique palindromes count: {len(unique_palindromes)}")
````

### Zadanie 3

Znajd藕 anagramy w pliku `english.txt`.

#### Co to jest anagram?
Anagramem sowa $s$ jest inne sowo utworzone poprzez przestawienie jego liter. Np. `pakt` i `ptak`.

#### Jak wykona zadanie?
* Krok 1: Umie wszystkie unikalne sowa skadajce si z wicej ni偶 jednego znaku w zbiorze.
* Krok 2: Utw贸rz sownik, gdzie kluczem s posortowane litery sowa, a wartoci jest lista wszystkich s贸w (z pliku `english.txt`), kt贸rych posortowane litery odpowiadaj kluczowi (s takim samym cigiem znak贸w).
* Krok 3: Wywietl wszystkie elementy sownika, kt贸re posiadaj wicej ni偶 jedno sowo na licie.

Na przykad dla s贸w: `alga`, `pakt`, `gala`, `ptak`, `lata`, wynikowy sownik powinien wyglda nastpujco:
```python
{
    'aagl': ['alga', 'gala'],
    'akpt': ['pakt', 'ptak'],
    # 'aalt': ['lata'] tego NIE wywietlamy
}
```

````python
with open("english.txt", "r", encoding="latin-1") as input_file:
    unique_words = set([word for word in input_file.read().split() if len(word) > 1])
    anagrams = {}
    for word in unique_words:
        sorted_word = ''.join(sorted(word))
        anagrams[sorted_word] = anagrams[sorted_word] + [word] if sorted_word in anagrams else [word]
    anagrams = {key: value for key, value in anagrams.items() if len(anagrams[key]) > 1}
    for anagram in anagrams:
        print(anagram, anagrams[anagram])
    print(f"anagrams count: {len(anagrams)}")
````

Do listy `seeds` umieszczamy 100 losowych liczb 64-bitowych.

4. Przyjmij niewielk warto `Q` (np. 15).
Dla ka偶dego cigu (tj. rozwa偶anego akapitu):

> Na wyjciu mamy zatem (dla 11 akapit贸w) 11 * 100 wartoci haszy.

5. Rozwa偶 pary akapit贸w ka偶dy z ka偶dym. Jeli dla danej pary co najmniej (np.) 30 haszy jest wsp贸lnych, to uwa偶amy akapity za podobne (by mo偶e plagiat) i wywietlamy na ekranie.

7. Poeksperymentuj z liczb u偶ytych funkcji haszujcych, wartoci `Q`, stopniem modyfikacji oryginalnych akapit贸w tekstu, progiem detekcji akapit贸w podobnych.

````python
import random
import numpy as np
import itertools
import time

def get_file_lines(file):
    with open(file, "r", encoding="utf-8") as input_file:
        lines = [line.rstrip() for line in input_file.readlines()]
        return [line for line in lines if line]

def get_min_hash(line, seed, q):
    hashes = []
    for end in range(q - 1, len(line)):
        start = end - q + 1
        hashes.append(hash(line[start:end + 1]) ^ seed)
    return min(hashes)

def print_similar_lines(file, seeds_count, q, limit):
    lines = get_file_lines(file)
    seeds = [random.getrandbits(64) for i in range(seeds_count)]
    lines_hashes = {line:np.array([get_min_hash(line, seed, q) for seed in seeds]) for line in lines}
    similar_lines = {}
    for line_hashes, compared_line_hashes in itertools.permutations(lines_hashes, 2):
        if len(np.intersect1d(lines_hashes[line_hashes], lines_hashes[compared_line_hashes])) >= limit:
            similar_lines[line_hashes] = similar_lines[line_hashes] + [compared_line_hashes] if line_hashes in similar_lines else [compared_line_hashes]
    for line in similar_lines:
        print(line)
        for similar_line in similar_lines[line]:
            print(similar_line)
        print()

file = "wikipedia.txt"
seeds_count = 200
q = 10
limit = 30

start = time.time()
print_similar_lines(file, seeds_count, q, limit)
end = time.time()
print(f"execution time: {(end - start):.2f} seconds")
````
